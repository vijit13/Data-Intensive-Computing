# Question 1: How does political views vary with age?

```r
pew= as.data.frame(pew)
```

```r
p1<-pew %>% select(ideo,age,race,sex)%>%filter(race==2 |race==3 |race==4 |race==5 )
p1<-as.data.frame(p1)
p1
```

```r
#p1$race
#p1$race[p1$race == 1]<-"White"
p1$race[p1$race==2]='Black'
p1$race[p1$race==3]='Asian'
p1$race[p1$race==4]='Mixed Race'
p1$race[p1$race==5]='Native American'
```

```r


p.dia <- ggplot(data=p1,aes(x=age))+geom_histogram(data=subset(p1,ideo==1),aes(x=age,y=..count..,fill='Very conservative'),binwidth=20,colour='white')+
geom_histogram(data=subset(p1,ideo==2),aes(x=age,y=..count..,fill='Conservative'),binwidth=20,colour='white')+
geom_histogram(data=subset(p1,ideo==3),aes(x=age,y=..count..,fill='Moderate'),binwidth=20,colour='white')+
geom_histogram(data=subset(p1,ideo==4),aes(x=age,y=..count..,fill='Liberal'),binwidth=20,colour='white')+
geom_histogram(data=subset(p1,ideo==5),aes(x=age,y=..count..,fill='Very Liberal'),binwidth=20,colour='white')+
facet_grid(. ~ race)

#p <- p.dia + layer(geom = "bar", mapping = aes(fill = ideo))
p.dia

```

# Hypothesis: Most of the people except Whites  have Moderate or Very Liberal political ideology. There are hardly any people who are Conservative in the political view .

# Question 3:

#  How many people think smartphone  is important when it comes to looking for job and career resources based on their emplyment status?

```r
p3<-pew %>% select(smjob3,emplnw,sex,age)%>%filter( !is.na(smjob3) & smjob3!=8 )%>% filter(emplnw==1 | emplnw==3 | emplnw==5 | emplnw==7)
p3
```

```r

p<- ggplot(data=p3,aes(x=smjob3))+geom_bar(data=subset(p3,emplnw==1),aes(x=smjob3,y=..count..,binwidth=1,fill='Employed Full-Time'))+
geom_bar(data=subset(p3,emplnw==3),aes(x=smjob3,y=..count..,binwidth=1,fill='Retired'))+
geom_bar(data=subset(p3,emplnw==5),aes(x=smjob3,y=..count..,binwidth=1,fill='Self-Employed'))+
geom_bar(data=subset(p3,emplnw==7),aes(x=smjob3,y=..count..,binwidth=1,fill='Student'))+
 labs(title = "", x = "Importance of using smartphone  for job and career resources", y = "Count")

p
```

## Importance Order:
### 1   Very important
### 2	Somewhat important
### 3	Not too important
### 4	Not at all important

# Hypothesis: Majority of full Time Employees  think that smartphone  is important when it comes to looking for job and career resources. Most of the Retired employees think that smartphone does not play a major role in finding job resources. Students also think that is plays a major role in getting jobs.



# Question 4:

#  How many people use dating sites based on their current marital status?

```r
p4<-pew %>% select(age,marital,date1a,date2a)%>%filter(!is.na(date1a) & !is.na(date2a))%>%filter( date1a==1 |date1a==2 |date2a==1 |date2a==2 )%>% filter(marital==1 )


```

```r
#library(ggplot2)
 ggplot(data = p4, aes(x = age))+
geom_histogram(data=subset(p4,date1a==2),aes(x=age,y=..count..,fill='Did not use Online Dating'),binwidth=5,colour='white', alpha = .5)+
geom_histogram(data=subset(p4,date1a==1),aes(x=age,y=..count..,fill='Used Online Dating'),binwidth=5,colour='white', alpha = .5)
```

# Hypothesis: Out of all the married people in the age group 25-75 only 10% used online dating sites. This shows that people are loyal to their partners and are not interested in exploring online dating options.

# Question 5: 

# How likely do people think  that in the next 50 years, robots and computers will do much of the work currently done by humans?

```r
p5<-pew %>% select(age,auto3)%>%filter(!is.na(auto3))%>% filter(auto3==1 | auto3==2 | auto3==3 | auto3==4 )
```

```r
#library(ggplot2)
 ggplot(data = p5, aes(x = age))+

geom_histogram(data=subset(p5,auto3==2),aes(x=age,y=..count..,fill='Probably Happen'),binwidth=5,colour='white', alpha = .5)+
geom_histogram(data=subset(p5,auto3==3),aes(x=age,y=..count..,fill='Probably NOT happen'),binwidth=5,colour='white', alpha = .5)+
geom_histogram(data=subset(p5,auto3==1),aes(x=age,y=..count..,fill='Definitely Happen'),binwidth=5,colour='white', alpha = .5)+
geom_histogram(data=subset(p5,auto3==4),aes(x=age,y=..count..,fill='Definitely NOT Happen'),binwidth=5,colour='white', alpha = .5)
```

# Hypothesis: Almost 50% of the people between the age of 35-75  think that robots will probably take over the work currently being done by humans.
