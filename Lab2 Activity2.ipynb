# Reading Team Table

```r
p6=tbl(my_db, sql("SELECT  team_long_name ,team.team_api_id , 
COALESCE(buildUpPlaySpeed,0) + COALESCE(buildUpPlayDribbling,0)+ COALESCE( buildUpPlayPassing,0) as BuildUpAll, 
COALESCE(chanceCreationPassing,0) + COALESCE(chanceCreationCrossing,0)+ COALESCE(chanceCreationShooting ,0) as ChanceUpAll,
COALESCE(defencePressure,0) + COALESCE(defenceAggression,0)+ COALESCE(defenceTeamWidth,0) as DefUpAll
                    from Team , Team_Attributes 
                    where Team.team_api_id=Team_Attributes.team_api_id  "))
collect(p6)

p6<-as.data.frame(p6)
p6[is.na(p6)] <- 0

```

```r
team <- p6 %>% 
             group_by(team_api_id) %>%
            filter(BuildUpAll == max(BuildUpAll))%>%
              select(team_long_name,BuildUpAll) %>%
              arrange(desc(BuildUpAll))

head(unique(team),10)
```

```r
team1 <- p6 %>% 
             group_by(team_api_id) %>%
            filter(ChanceUpAll == max(ChanceUpAll))%>%
              select(team_long_name,ChanceUpAll) %>%
              arrange(desc(ChanceUpAll))

head(unique(team1),10)
```

```r
team3 <- p6 %>% 
             group_by(team_api_id) %>%
            filter(DefUpAll == max(DefUpAll))%>%
              select(team_long_name,DefUpAll) %>%
              arrange(desc(DefUpAll))

head(unique(team3),10)
```

```r
p7=tbl(my_db, sql("SELECT home_team_api_id,sum(home_team_goal) from match group by home_team_api_id"))
collect(p7)
p8=tbl(my_db, sql("SELECT away_team_api_id,sum(away_team_goal) from match group by away_team_api_id"))
collect(p8)
p7<-as.data.frame(p7)
p8<-as.data.frame(p8)
p7[is.na(p7)] <- 0
p7[is.na(p8)] <- 0
names(p7)[names(p7)=="home_team_api_id"] <- "team_api_id"
names(p8)[names(p8)=="away_team_api_id"] <- "team_api_id"
names(p7)[names(p7)=="sum(home_team_goal)"] <- "GoalTotal"
names(p8)[names(p8)=="sum(away_team_goal)"] <- "GoalTotal"

p9=rbind(p7,p8,all=TRUE)
#p9
NumberOfGoals<-aggregate(GoalTotal~., p9, FUN=sum)%>%  arrange( desc(GoalTotal))
#NumberOfGoals
```
