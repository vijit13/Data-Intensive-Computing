# Filtering relevant coulmns from the dataset

```r
newdf<-df %>% select(age,sex,race,inc,emplnw) %>%filter(!(race %in% c("6","8","9")) & !(inc %in% c("98","99")) & !(emplnw %in% c("8","98","99")) )
```

```r
newdf
```

# Normalizing the values for KMeans

```r
nrow(newdf)

newdf$age<-((newdf$age - min(newdf$age)) / (max(newdf$age) - min(newdf$age)))
newdf$sex<-((newdf$sex - min(newdf$sex)) / (max(newdf$sex) - min(newdf$sex)))
newdf$race<-((newdf$race - min(newdf$race)) / (max(newdf$race) - min(newdf$race)))
newdf$inc<-((newdf$inc - min(newdf$inc)) / (max(newdf$inc) - min(newdf$inc)))
newdf$emplnw<-((newdf$emplnw - min(newdf$emplnw)) / (max(newdf$emplnw) - min(newdf$emplnw)))
```

```r
newdf$race
```

 # Kmeans with 3 Clusters

```r
set.seed(20)
pewCluster <- kmeans(newdf, 3, nstart = 20)
pewCluster
```

```r
table(pewCluster$cluster, newdf$inc)
pewCluster$size
```

```r
plot(newdf[c("age", "emplnw")], col=pewCluster$cluster)
points(pewCluster$centers[,c("age", "emplnw")], col=1:3, pch=23, cex=3)
```
