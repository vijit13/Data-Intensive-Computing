# B. 
### Change the model so that it line is forced through Wayne Gretzky. Note the R2 and p values of the fit. This will another line or if you prefer another plot

```r
wayne<-lm(I(A-1963) ~ I(G-894) , data=nhl)
summary(wayne)
coef(wayne)
plot(A~G , data=nhl)

intercept=1963-0.5202*894
abline(intercept,0.5202,col="red")

n2=data.frame(c("b"),c(0.09913),c(0.0008317))
```

```r
colnames(n2)<-c("Model","R Squared Value","p-Value")
```

# D.
## Make all the model go through (0,0) and create new models. Note R2 and p values of the fit

```r
nhllm0 <- lm(A ~ G+0, data = nhl)
summary(nhllm0)
coef(nhllm0)
plot(A~G , data=nhl)
abline(nhllm0,col="red")


n4=data.frame(c("Normal LM with origin"),c(0.8975),c("2.2e-16"))
colnames(n4)<-c("Model","R Squared Value","p-Value")
```

```r
wayne0<-lm(I(A-1963) ~ I(G-894)+0 , data=nhl)
summary(wayne0)
coef(wayne0)
plot(I(A-1963) ~ I(G-894), data=nhl)

#slope=1963/894
abline(wayne0,col="red")


n5=data.frame(c("Wayne with Origin"),c(0.9094),c("2.2e-16"))
colnames(n5)<-c("Model","R Squared Value","p-Value")
```

```r
kane0<-lm(I(A-465) ~ I(G-285)+0 , data=nhl)
summary(kane0)
coef(kane0)
plot(I(A-465) ~ I(G-285) , data=nhl)

#slope=465/285
abline(kane0,col="red")

n6=data.frame(c("Patrick with Origin"),c(0.5689),c("2.2e-16"))
colnames(n6)<-c("Model","R Squared Value","p-Value")
```
